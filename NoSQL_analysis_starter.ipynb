### 1. Which establishments have a hygiene score equal to 20?

````python
# Assign the collection to a variable
establishments = db['establishments']

# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
result = establishments.find(query)

# Use count_documents to display the number of documents in the result
num_documents = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {num_documents}")

# Display the first document in the results using pprint
if num_documents > 0:
    first_document = result[0]
    print("First document with a hygiene score of 20:")
    pprint(first_document)
else:
    print("No establishments found with a hygiene score of 20.")
````

````python
# Convert the result to a Pandas DataFrame
import pandas as pd
df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
````

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

````python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": {"$regex": "London"},  # Case-insensitive regex
    "RatingValue": {"$gte": "4"}  
}

# Use count_documents to display the number of documents in the result
result_count = establishments.count_documents(query)
print(result_count)

# Display the first document in the results using pprint
result_cursor = establishments.find(query)
first_document = result_cursor.next() if result_count > 0 else None

if first_document:
    print("First document in the results:")
    pprint(first_document)
else:
    print("No documents found.")
````

````python
import pandas as pd

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
df.head(10)
````

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

````python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": "0"}},
    {"$group": {"_id": "$LocalAuthorityName", "totalcount": {"$sum": "1"}}},
    {"$sort": {"totalcount": -1}}
]

result = establishments.aggregate(pipeline)

# Print the number of documents in the result
num_documents_pipeline = sum(1 for _ in result)
print(f"Number of documents in the result of the pipeline: {num_documents_pipeline}")

# Print the first 10 results
result = establishments.aggregate(pipeline)  # Need to re-run the pipeline since it's consumed in the previous operation
print("First 10 results of the pipeline:")
for document in result:
    pprint(document)
````

````python
# Convert the result to a Pandas DataFrame
df_pipeline_results = pd.DataFrame(result)

# Display the number of rows in the DataFrame
num_rows_pipeline = len(df_pipeline_results)
print(f"Number of rows in the DataFrame for pipeline results: {num_rows_pipeline}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame for pipeline results:")
print(df_pipeline_results.head(10))
````
